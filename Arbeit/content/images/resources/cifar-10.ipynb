```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10

plt.style.use("seaborn-whitegrid")
plt.rc("font", size=14)
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
labels = [
    "Flugzeug", "Automobil", 
    "Vogel", "Katze", 
    "Hirsch", "Hund", 
    "Frosch", "Pferd", 
    "Schiff", "Lastkraftwagen"
]
```

```python
img_per_class = 1
img_label_pairs = list(zip(X_train, map(lambda x: x[0], y_train)))

imgs = {
    label : [ img for img, j in img_label_pairs if i == j ][:img_per_class]
    for i, label in enumerate(labels)
}
```

```python
fig, axs = plt.subplots(2, 5, figsize=(10, 4), constrained_layout=True)
axs = axs.reshape((-1,))

for i, label in enumerate(labels):
    for ax, img in zip(axs[i*img_per_class:img_per_class+i*img_per_class], imgs[label]):
        ax.imshow(img)
        ax.set_title(label)
        ax.set_axis_off()
        
plt.show()
```

```python
fig.savefig("cifar-10.pdf", bbox_inches='tight')
```
