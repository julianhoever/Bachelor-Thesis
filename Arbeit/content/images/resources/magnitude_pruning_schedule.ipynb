```python
import matplotlib.pyplot as plt

plt.style.use("seaborn-whitegrid")
plt.rc("font", size=14)
```

```python
def s_t(s_i, s_f, n, t_0, delta_t, t):
    return s_f + (s_i - s_f) * (1 - (t - t_0)/(n * delta_t))**3
```

```python
s_i = 0
s_f = 80
n = 6
t_0 = 0
delta_t = 5


Xs = [ t for t in range(t_0 + n * delta_t + 1) ]
Ys = [ 0 if t < t_0 else s_t(s_i, s_f, n, t_0, delta_t, t_0 + ((t - t_0) // delta_t) * delta_t) for t in Xs ]

fig, ax = plt.subplots(1, 1, figsize=(12, 5))
ax.step(x=Xs, y=Ys, where="post")
ax.set_xlabel("Anzahl der Trainingsschritte")
ax.set_ylabel("Sparsity [%]")

plt.show()
```

```python
fig.savefig("magnitude_pruning_schedule.pdf", bbox_inches='tight')
```
